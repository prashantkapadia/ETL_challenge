```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Store CSV into DataFrame
csv_file = "Resources/netflix_titles.csv"
netflix_data_df = pd.read_csv(csv_file)
netflix_data_df.head()
```

```python
# Storing IMDB_movies.csv aquired from Data World.
imdb_file = "Resources/IMDb_movies.csv"
imdb_df = pd.read_csv(imdb_file)
imdb_df.head(5)
```

```python
# Create a filtered dataframe from specific columns

imdb_mov = ["imdb_title_id", "title", "year", "genre", "duration", "country", "director", "actors",
            "budget", "usa_gross_income", "worlwide_gross_income", "reviews_from_users"]
imdb_trnfmd = imdb_df[imdb_mov].copy()

# Rename the column headers
imdb_trnfmd = imdb_trnfmd.rename(columns={"imdb_title_id": "title_id",
                                                          "usa_gross_income": "gross_income_usa",
                                                          "worlwide_gross_income": "gross_income_global",
                                                          "reviews_from_users": "reviews"})

# Clean the data by dropping duplicates

imdb_trnfmd.drop_duplicates("title_id", inplace=True)

# Filtering DF by keeping based on year 2000 and up, and Country = 'USA'
imdb_trnfmd = imdb_trnfmd.loc[imdb_trnfmd["year"] >= 2000, :]
imdb_trnfmd = imdb_trnfmd.loc[imdb_trnfmd["country"] == "USA", :]

# Copy of existing dataframe
imdb_movies_df = imdb_trnfmd.copy()

#Dropping NaN Values from whole dataframe

imdb_movies_df = imdb_movies_df.dropna()
```

```python
# This is commented because we ran this block of code successfully
# Removing $, EUR, AD currency indicators and turn columns into Integer from Object.
imdb_movies_df[imdb_movies_df.columns[9:11]] = imdb_movies_df[imdb_movies_df.columns[9:11]].replace('[$,EUR,AD]', '', regex=True).astype(float)
```

```python
# Removing $,AUD, ER currency operator from "budget" column.
imdb_movies_df[imdb_movies_df.columns[8]] = imdb_movies_df[imdb_movies_df.columns[8]].replace('[$,AUD,ER]', '', regex=True).astype(float)
```

```python
# Droping the country column
imdb_movies_df.drop(columns=['country'],inplace = True)
```

```python
#imdb_movies_df.sort_values(by=['year'])
imdb_movies_df.sort_values(by=['title', 'year'], ascending= True, inplace = True )
# Resetting index
imdb_movies_df.reset_index()
```

```python
#Extracting only 'Movies' Type (not Series)
only_movies = netflix_data_df.loc[netflix_data_df["type"] == "Movie", :]
#Extracting only 'country 'United States'
only_us_df = only_movies.loc[only_movies["country"] == "United States", :]
# Dropping NaN Values 
clean_df = only_us_df.dropna()
# Filtering only new movies after year 2000

clean_df = clean_df.loc[clean_df["release_year"] >= 2000, :]
# Reorganizging columns 
clean_df = clean_df[['show_id','type','title','director','cast','release_year','rating','duration','listed_in','description']]
# Removing "min" from "duration" column value
clean_df[clean_df.columns[7]] = clean_df[clean_df.columns[7]].replace('[min]', '', regex=True).astype(float)
# Sorting and organizing DF based on Title and release year.
clean_df.sort_values(by=['title', 'release_year'], ascending= True, inplace = True )

# Dropping column "Movie" due to repeated value
clean_df.drop(columns=['type'],inplace = True)
# Reseting index
clean_df.reset_index(inplace = True)
```

```python
# CREATING COPY OF EXISTING DF AS SAFE DATA POINT FOR ANY FUTURE CHANGES.
netflix_clean_df = clean_df.copy()
```

```python
# changing column name cast conflicting with pgadmin built in function.
netflix_clean_df = netflix_clean_df.rename(columns={"cast": "actors"})

```

```python
# Dropping column name showed up in DF conflicting with PGADMIN DB.
netflix_clean_df.drop(columns=['index'],inplace = True)
```
